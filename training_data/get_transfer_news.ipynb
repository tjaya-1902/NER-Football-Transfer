```python
import requests
from bs4 import BeautifulSoup

url = "https://www.bbc.com/sport/football/articles/cgkyg77pm30o"
response = requests.get(url)
soup = BeautifulSoup(response.text, "html.parser")

# 1. Get headline
headline = soup.find("h1", id="main-heading")
print("Headline:", headline.get_text(strip=True) if headline else "Not found")

# 2. Get bolded lead paragraph (typically the key sentence)
lead_para = soup.find("b", class_="ssrcss-1xjjfut-BoldText")
print("Lead Sentence:", lead_para.get_text(strip=True) if lead_para else "Not found")
```
